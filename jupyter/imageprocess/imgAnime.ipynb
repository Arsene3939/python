```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math
x=np.linspace(0,4*np.pi,1000)
sinx=np.sin(x)
squareX=np.sign(sinx)
fig,ax=plt.subplots(figsize=(5,5))
setup=[['left','zero'],['right','zero'],['top','zero'],['bottom','zero']]
for i in setup:
    ax.spines[i[0]].set_position(i[1])


R1=1
R2=4
beta=R1/R2
Vcc=15
Vx=Vcc
C=10E-6
R=1000
VU=Vcc*beta
VL=-VU
Vc=[0]
Vo=[0]
V1=[15]
T=4*R*C*beta
t=np.linspace(0,3*T,1000)
for i in range(len(t)):
    if -Vc[i] > VU:
        Vx=Vcc
    elif -Vc[i] < VL:
        Vx=-Vcc
    Vo.append(-Vc[i])
    Vc.append(Vc[i]+((Vx/R)*t[1])/C)
    V1.append(Vx)
dot1,=ax.plot(t[0],Vc[1],'o',color='red')
dot2,=ax.plot(t[0],Vc[1],'o',color='red')
ax.plot(t,Vc[1:],t,V1[1:])
def anime(i):
    dot1.set_data(t[i],Vc[i])
    dot2.set_data(t[i],V1[i])
    return dot1,dot2
animegif=animation.FuncAnimation(fig, anime, frames=len(t), interval=10)
```
