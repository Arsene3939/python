```python
from random import *
s="5123456789+-*/"
ls=[]
for i in range(20):
    seq=""
    flag=1
    t=randint(0,14)
    n=0
    for i in range(randint(20,35)):
        if t>9:
            flag=1
        else:
            n+=1
        if flag:
            t=randint(0,9)
            flag=0
            n+=1
        elif n>=3:
            t=randint(10,13)
            n=0
        else:
            t=randint(0,13)
            
        seq+=s[t]
    if seq[-1] in "+-*/":
        seq=seq[:-1]
    ls.append(seq)
for i in ls:
    print(i)
print()
for i in ls:
    eval(i)
    print(i)
    print("%.10f"%eval(i))
```

```python
import random
ls=[]
for x in range(20):
    num=[]
    for i in range(4):
        r=random.randint(-100000,100000)/1000
        if r<0:
            num.append("(%s)"%str(r))
        else:
            num.append(str(r))
    opr=list("*")
    for i in range(len(num)-1):
        num[0]+=opr[0]+num[1]
        if(random.randint(0,3)==1):
            num[0]="(%s)"%num[0]
        num.pop(1)
        random.shuffle(opr)
        random.shuffle(num)
    ls.append(num[0])
for i in ls:
    print(i)
print()
for i in ls:
    print(i)
    try:
        print("%.10f"%eval(i))
    except Exception as e:
        print("error ",end="")
        print(e)
```
