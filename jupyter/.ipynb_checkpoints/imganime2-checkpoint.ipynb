```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib nbagg
sort=[1,2,3,4,1,1,0,0,0]
img=[]
for i in sort:
    img.append(plt.imread(str(i)+".jpg"))
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=(2,1))
photo=ax.imshow(img[0])
def anime(i):
    photo.set_data(img[i])
    return photo
aim=animation.FuncAnimation(fig,anime,frames=len(img),interval=500)
```
