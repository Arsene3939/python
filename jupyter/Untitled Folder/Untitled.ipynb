```python
import random as r
import numpy as np
import time
import matplotlib.pyplot as plt
def quick_sort(partition,arr,low,high):
    if(low<high):
        pivot_ind=partition(arr,low,high)
        quick_sort(partition,arr,low,pivot_ind)
        quick_sort(partition,arr,pivot_ind+1,high)
def hoare(arr,low,high):
    pivot=arr[low]
    left=low-1
    right=high+1
    while(1):
        
        while(1):
            left+=1
            if arr[left]>=pivot:
                break
        while(1):
            right-=1
            if arr[right]<=pivot:
                break
        if(left>=right):
            return right
        arr[left],arr[right]=arr[right],arr[left]
def lomuto(arr,low,high):
    pivot=arr[high]
    left =low-1
    for i in range(low,high):
        if arr[i]<=pivot:
            left+=1
            arr[i],arr[left]=arr[left],arr[i]
    
    arr[high],arr[left+1]=arr[left+1],arr[high]
    return left
def timeof(partition,run_n,length):
    time1=time.perf_counter()
    for i in range(run_n):
        arr=[str(r.randint(0,100))*100 for x in range(length)]
        quick_sort(hoare,arr,0,len(arr)-1)      
    return time.perf_counter()-time1

x=[i for i in range(100,1001,100)]
y1=[timeof(lomuto,100,i)for i in x]
y2=[timeof(hoare,100,i)for i in x]
plt.plot(x,y1,x,y2)
```
