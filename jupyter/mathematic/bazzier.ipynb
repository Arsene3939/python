```python
import matplotlib.pyplot as plt
import numpy as np
def B(t,p):
    n=len(p)
    if n==1:
        return p[0]
    return B(t,p[1:]*t)+B(t,p[:-1]*(1-t))
fig,ax=plt.subplots(figsize=(10,5))
px=np.array([2,3,2,4,2,2,2,2,2,2,2])
x=[i for i in np.linspace(0,1,100)]
y=[B(i,px)for i in np.linspace(0,1,100)]
ax.plot(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from random import *
def B(t,p): 
    if t==1:return p[-1] 
    c=1 
    v=0 
    n=len(p)-1
    t_=t/(1-t) 
    tt=(1-t)**n
    for i in range(n+1): 
        v+=c*tt*p[i] 
        tt*=t_
        c*=(n-i)/(i+1) 
    return v
fig,ax=plt.subplots(figsize=(10,5))
px=np.array([2,3,2,4,2,2,2,2,2,2,2])
x=[i for i in np.linspace(0,1,100)]
y=[B(i,px)for i in np.linspace(0,1,100)]
ax.plot(x,y)
```
