```python
import os,sys
import shutil
import matplotlib.pyplot as plt
cd="C:\\Users\\ASUS\\AppData\\Local\\osu!\\songs"
allt=len(os.listdir(cd))
i=0
for song in os.listdir(cd):
    print("%d/%d"%(i,allt))
    i+=1
    #print(song)
    j=0
    for img in os.listdir(cd+"\\"+song):
        if(len(img.split('.'))==1):
            continue
        subfilename=img.split('.')[1]
        if(subfilename=="png"or subfilename=="jpg"or subfilename=="jiff"or subfilename=="bmp"):
            try:
                print(cd+"\\"+song+"\\"+img)
                arr=plt.imread(cd+"\\"+song+"\\"+img)
                if(len(arr)*len(arr[0])>350000):
                    shutil.copyfile(cd+"\\"+song+"\\"+img,"D:\\_self_\\%s%d.%s"%(song,j,subfilename))
                    j+=1
            except:
                shutil.copyfile(cd+"\\"+song+"\\"+img,"D:\\_self_\\%s%d.%s"%(song,j,subfilename))
            #print("     ",img)
```
