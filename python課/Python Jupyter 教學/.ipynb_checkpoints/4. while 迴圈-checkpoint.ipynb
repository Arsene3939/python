# 4. while 迴圈

### 迴圈(或稱重複處理)是處理大量資料必須的功能，是程式設計不可或缺的基礎。

### Python提供for 與while兩種迴圈指令。for 使用於迭代物件的迴圈處理，while使用於條件式迴圈。


### while 語法

#### while 條件式 :  
    指令1  
    指令2  
    其他指令  

#### 條件式成立時，重複執行迴圈內部指令，直到條件式不成立為止

## 連續印出 n 個數字

```python
i = 0 #設定迴圈變數的起始值
while i < 10: #檢查迴圈變數是否符合條件
    print(i)
    i += 1 #更新迴圈變數，迴圈變數不更新時可能造成無窮迴圈
    
```

## 迴圈設計不當可能形成無窮迴圈，將造成CPU過度耗用拖累系統效能

### 解決方法
### 1. 中斷或重啟系統核心(kernel)
### 2. 更改程式錯誤

```python
i = 1 #設定迴圈變數的起始值
while i <= 10: #檢查迴圈變數是否符合條件
    print(i)
    #i += 1 #更新迴圈變數，迴圈變數不更新時可能造成無窮迴圈
    
```

### 計算 $1^2 + 2^2 + 3^2 + ... + 9^2 + 10^2 = ?$

```python
i=1
sum=0
while i<= 10:
    sum += i**2   
    i += 1
print(sum)
```

### 從鍵盤輸入10個整數(一次一個)，計算其和。

```python
i=1
sum=0
while i<= 10:
    n = int(input())
    sum += n   
    i += 1
print(sum)
```

## break中斷指令	
while 條件:  
  &emsp;程式碼區塊1  
  &emsp;if 條件運算式:  
    &emsp;&emsp;程式區塊2  
    &emsp;&emsp;break  
  &emsp;程式區塊3  		

### 說明:
1. 當x使條件運算式不成立時，   
    迴圈依序執行程式碼區塊1、程式碼區塊3，繼續迴圈的下一個元素  
2. 當x使條件運算式成立時，  
    迴圈依序執行程式碼區塊1、程式碼區塊2，中斷迴圈


```python
#尋找滿足 1+2+3+…+N >100 的最小N值
sum = 0
i = 1
while True:
    sum += i
    print(i, sum)
    if sum>100:        
        break
    i += 1
```

## continue繼續指令	
while 條件:  
  &emsp;程式碼區塊1  
  &emsp;if 條件運算式:  
    &emsp;&emsp;程式區塊2  
    &emsp;continue  
  &emsp;程式區塊3		

### 說明
1. 當x使條件運算式不成立時，  
    迴圈依序執行程式碼區塊1、程式碼區塊3，繼續迴圈的下一個元素  
2. 當x使條件運算式成立時，  
    迴圈依序執行程式碼區塊1、程式碼區塊2，繼續迴圈的下一個元素  


```python
# 印出1~20中，非3倍數的數
i = 0
while i < 20:
    i += 1
    if i%3 == 0:
        continue
    print(i)
```
