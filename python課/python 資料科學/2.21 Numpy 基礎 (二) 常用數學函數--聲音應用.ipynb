## 畫正弦波波形

### 正弦波的一般式 $s(t) = A\sin (2 \pi ft + \theta)$

A: 波形的振幅  
f: 波形的頻率  
$\theta $ : 波形的相位

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 2, 0.01)
A = 10
f = 1
theta = np.pi / 3

s = A * np.sin(2*np.pi*f*t + theta)

fig, ax = plt.subplots(figsize=(8,8))
ax.plot(t, s, label='s')


ax.set(xlabel='t',ylabel='value',title='Capacitor charging')
ax.legend()
```

## 聽正弦波的聲音

## 固定頻率

```python
import numpy as np
import matplotlib.pyplot as plt
from Audio import Audio #載入聲音播放程式

samplingRate = 22050  #取樣頻率
M = 5  #聲音秒數
totalPoints = samplingRate * M
t = np.linspace(0,M,totalPoints)
f=440  #聲音頻率

data = 1000*np.sin(2*np.pi*f*t)
plt.plot(data[:1000]) #取前1000筆資料
plt.show()
Audio(data,rate=samplingRate)
```

## 線性增頻

```python
import numpy as np
import matplotlib.pyplot as plt
#import Audio.Audio
samplingRate = 22050  #取樣頻率
M = 5  #聲音秒數
totalPoints = samplingRate * M
t = np.linspace(0,M,totalPoints)
f= np.linspace(100,2500,totalPoints)  #聲音頻率
data = 1000*np.sin(2*np.pi*f*t)
#plt.plot(data[:1000]) #取前1000筆資料
plt.plot(t,f)
plt.show()
Audio(data,rate=samplingRate)
```

## 指數增頻

```python
import numpy as np
import matplotlib.pyplot as plt
#import Audio.Audio
samplingRate = 22050  #取樣頻率
M = 5  #聲音秒數
totalPoints = samplingRate * M
t = np.linspace(0,M,totalPoints)
f= np.logspace(2,3.5,totalPoints)  #聲音頻率
data = 1000*np.sin(2*np.pi*f*t)
#plt.plot(data[:1000]) #取前1000筆資料
plt.plot(t,f)
plt.show()
Audio(data,rate=samplingRate)
```
