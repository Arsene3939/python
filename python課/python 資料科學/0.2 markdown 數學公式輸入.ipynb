# 插入數學公式

## 在Markdown中插入數學公式的語法是 
    
    行內公式：使用兩個 $號引用公式: $数学公式$

    獨立公式：使用兩對$$符號引用公式：$$数学公式$$
    

行內公式是可以讓公式在文中與文字或其他東西混編，不獨占一行。

示例 

質能方程 \\$E = mc^2\\$
    
顯示

質能方程$E = mc^2$



獨立公式使公式單獨占一行，不與文中其他文字等混編。

示例
    質能方程\\$\\$E = mc^2\\$\\$
    
顯示
    質能方程$$E = mc^2$$



## 普通公式
    普通的加減乘除數學公式的輸入方法與平常的書寫一樣。

示例
\\$\\$x = 100 * y + z - 10 / 33 + 10 % 3\\$\\$

顯示

$$x = 100 * y + z - 10 / 33 + 10 % 3$$


## 上下標

使用^來表示上標，_來表示下標，同時如果上下標的內容多於一個字符，可以使用{}來將這些內容括起來當做一個整體。
與此同時，上下標是可以嵌套的。

示例
\\$\\$x = a _{1}^n + a_ {2}^n + a_{3}^n\\$\\$
顯示

$$x = a_{1}^n + a_{2}^n + a_{3}^n$$

如果希望左右兩邊都能有上下標，可以使用\sideset語法

示例

\\$\\$\sideset{^1 _2}{^3_ 4}A\\$\\$

顯示

$$\sideset{^1_2}{^3_4}A$$

## 括號

    ()，[]和|都表示它們自己，但是{}因為有特殊作用因此當需要顯示大括號時一般使用\lbrace \rbrace來表示。

示例

\\$\\$f(x, y) = 100 * \lbrace[(x + y) * 3] - 5\rbrace\\$\\$

顯示

$$f(x, y) = 100 * \lbrace[(x + y) * 3] - 5\rbrace$$



## 分數

    分數使用\frac{分母}{分子}這樣的語法，不過推薦使用\cfrac來代替\frac，顯示公式不會太擠。

示例

\\$\\$\frac{1}{3} 與\cfrac{1}{3}\\$\\$

顯示

$$\frac{1}{3} 與\cfrac{1}{3}$$



## 開方
    開方使用\sqrt[次数]{被开方数}這樣的語法

示例

\\$\\$\sqrt[3]{X}\\$\\$ 

\\$\\$\sqrt{5 - x}\\$\\$

顯示

$$\sqrt[3]{X}$$

$$\sqrt{5 - x}$$
