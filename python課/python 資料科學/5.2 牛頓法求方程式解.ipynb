# 5.2 牛頓法求方程式解

$如果x_n是f(x)=0的一個似近解，且f'(x) \neq 0, 則下一個近似解是$
$$x_{n+1}=x_n − \cfrac{f(x_n )}{f′(x_n )}$$


推導

$假設我們想尋找f(x)= 0的近似解，且已設定一個近似的起始解x_0 。$  
$首先找到f(x)在x_0 的切線 (下圖中藍色的線)$
$$y=f(x_0 )+f′(x_0 )(x−x_0 )$$
<img align="center" style="padding-right:10px;" src="figures\newton_method.png">

$此圖顯示切線與x軸的交點x_1 比x_0 更靠近正確解(上圖中x軸上的紅點)，我們可使用此點當作新的近似解。$
$x_1 求解方法如下,$
$$0=f(x_0 )+f′(x_0 )(x_1 −x_0 )$$

$$x_{1} - x_0 = − \cfrac{f(x_0 )}{f′(x_0 )}$$
$$x_{1}=x_0 − \cfrac{f(x_0 )}{f′(x_0 )}$$

$同理，我們可以使用x_1 求另一個更靠近的近似解x_2 $
$$x_{2}=x_1 − \cfrac{f(x_1 )}{f′(x_1 )}$$

以此類推，新近似解的迭代表達式為
$$x_{n+1}=x_n − \cfrac{f(x_n )}{f′(x_n )}$$



## 求解極值問題
牛頓法也被用於求函數的極值。由於函數取極值的點處的導數值為零，故可用牛頓法求導函數的零點，其迭代式為

$$x_{n+1}=x_n − \cfrac{f'(x_n )}{f''(x_n )}$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3.4,3.5,0.01)

y = x**5 -4*x**4 + 3*x**3 - 4*x**2 + 5 * x - 10
dy = 5*x**4 - 16*x**3 + 9 * x**2 - 8 *x + 5


plt.plot(x,x*0,x,y)
i=1
s = [3.5]
s2=[3.5]
for i in range(150):
    x = s[i-1]
    y = x**5 -4*x**4 + 3*x**3 - 4*x**2 + 5 * x - 10
    dy = 5*x**4 - 16*x**3 + 9 * x**2 - 8 *x + 5
    s.append(x - 0.2*y/dy)
    s2.append(x - 0.00001*y*dy)
    #print(s[i])
    #plt.plot(s[i],0,'*')

x = np.array(s)
y = x**5 -4*x**4 + 3*x**3 - 4*x**2 + 5 * x - 10
dy = 5*x**4 - 16*x**3 + 9 * x**2 - 8 *x + 5
plt.plot(x,y,'.')
x = np.array(s2)
y = x**5 -4*x**4 + 3*x**3 - 4*x**2 + 5 * x - 10
dy = 5*x**4 - 16*x**3 + 9 * x**2 - 8 *x + 5
plt.plot(x,y,'*')    
#s2
```
