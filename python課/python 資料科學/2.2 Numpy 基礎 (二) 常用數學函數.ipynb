## 常用數值產生函數

## arange

###      語法  np.arange( 起始值, 結束值, 步幅, 資料型別 )：

    產生一維陣列，和np.array()的差別在於arange擁有較大的彈性，而且元素數值是自動化產生！

    步幅可為小數，決定每隔多少數值產生一個元素。

```python
import numpy as np
np.arange(0,1,0.1)
```

## linspace

### 語法  np.linspace( 起始值, 結束值, 起始與結束的區間內要產生幾個元素 )：

    只要給定陣列的區間(起始值、結束值)，就可以要求在這個區間內產生幾個元素！

```python
np.linspace(0,1,5,endpoint=False)
```

## logspace

### 語法  np.logspace( 起始值, 結束值, 起始與結束的區間內要產生幾個元素 )：

    只要給定陣列的區間(起始值、結束值)，就可以要求在這個區間內產生幾個元素！

```python
np.logspace(1,2,5)
```

## 常用數學函數

    np.exp(a)  自然指數，以e為底
    np.log(a)  自然對數，以e為底
    np.sqrt(a) 平方根函數
    np.sin(a)  正弦函數
    np.cos(a)  餘弦函數
    
    函數作用於陣列的每一元素


```python
import numpy as np
a = np.arange(1,5,0.5)
print('exp:', np.exp(a))
print('log:', np.log(a))
print('sqrt:', np.sqrt(a))
print('sin:', np.sin(a))
print('cos:', np.cos(a))
```

## 畫各種函數圖

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.1, 2, 0.1)
exp = np.exp(x)
log = np.log(x)
sqrt = np.sqrt(x)
x2 = x ** 2

fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, exp, label='exp(x)')
ax.plot(x, sqrt, label='$\sqrt{x}$')
ax.plot(x, x, label='x')
ax.plot(x, x2, label='$x^2$')


ax.plot(x, log, label='log(x)')
ax.set(xlabel='x',ylabel='value',title='function value')
ax.legend()
```

## 練習: 畫出充放電曲線

### 補上電阻上的電壓曲線，適當使用圖例

```python
import numpy as np
import matplotlib.pyplot as plt


E = 10
R = 1
C = 0.1
tau = R * C 
#t = np.arange(0, 1, 0.01)
t = np.linspace(0, 5*tau, 100)
Vc = E * (1 - np.exp(-t/tau))

fig, ax = plt.subplots(figsize=(8,8))
ax.plot(t, Vc, label='Vc')


ax.set(xlabel='t',ylabel='value',title='Capacitor charging')
ax.set_xticks([0, tau, 2*tau,3*tau,4*tau,5*tau])
ax.set_xticklabels(['0',r'$\tau$',r'$2\tau$',r'$3\tau$',r'$4\tau$',r'$5\tau$'])
ax.legend()
```

## 畫正弦波波形

### 正弦波的一般式 $s(t) = A\sin (2 \pi ft + \theta)$

A: 波形的振幅  
f: 波形的頻率  
$\theta $ : 波形的相位

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 2, 0.01)
A = 10
f = 1
theta = np.pi / 3

s = A * np.sin(2*np.pi*f*t + theta)

fig, ax = plt.subplots(figsize=(8,8))
ax.plot(t, s, label='s')


ax.set(xlabel='t',ylabel='value',title='Capacitor charging')
ax.legend()
```

### 練習: RLC 電路的電壓、電流波形圖

<img align="center" style="padding-right:10px;" src="figures\RLC_Circuit.png">

$\omega = 2 \pi f$   
$e(t)=\sqrt{2}E\sin(\omega t + \theta_E)$  
$\vec{E} = E \angle \theta_E $  

$\vec{X_L}=X_L\angle \theta_L, X_L = \omega L, \theta_L = 90°$  
$\vec{X_C}=X_C\angle \theta_C, X_C = \cfrac{1}{\omega C}, \theta_C = -90°$  
$\vec{X} = \vec{X_L} - \vec{X_C} =X\angle \pm 90°$  
$\vec{Z} = Z \angle \theta_Z,  Z=\sqrt{R^2 + X^2}, \theta_L = \arctan \cfrac{X}{R}$  
$\vec{I} = \cfrac{\vec{E}}{\vec{Z}} = I \angle \theta_I, I = \cfrac{E}{Z}, \theta_I = \theta_E - \theta_Z$  
$\vec{V_R} =\vec{I} R =  V_R \angle \theta_{V_R}, V_R = IR, \theta_{V_R} = \theta_I$  
$\vec{V_L} =\vec{I} \vec{X_L} = V_L \angle \theta_{V_L}, V_L = IX_L , \theta_{V_L} = \theta_I + 90°$  
$\vec{V_C} =\vec{I} \vec{X_C} = V_C \angle \theta_{V_C}, V_C = IX_C , \theta_{V_C} = \theta_I - 90°$


$e(t)=\sqrt{2}E\sin(\omega t + \theta_E)$  
$i(t)=\sqrt{2}I\sin(\omega t + \theta_I)$  
$v_R (t)=\sqrt{2}V_R \sin(\omega t + \theta_{V_R})$  
$v_L (t)=\sqrt{2}V_L \sin(\omega t + \theta_{V_L})$  
$v_C (t)=\sqrt{2}V_C \sin(\omega t + \theta_{V_C})$

### 練習: 

如上圖，R=10Ω，L=25mH, C=50µF ，$e(t)=100\sqrt{2}\sin(1000t)$，  
畫出各元件上的電壓波形(2個周期)，並加上適當圖例。

```python
import numpy as np
import matplotlib.pyplot as plt
R = 10
L = 25E-3
C = 50E-6
E = 100
theta_E = 0

w = 1000
f = w /(2 * np.pi)  #求頻率
T = 1 / f           #求週期

t = np.linspace(0,2*T, 200)  #2個週期，等分200份

XL = w * L
Xc = 1/(w * C)
X = XL - Xc
Z = (R**2 + X**2)**0.5
theta_Z = np.arctan(X/R)

I = E / Z
theta_I = theta_E - theta_Z

VR = I * R
theta_VR = theta_I

VL = I * XL
theta_VL = theta_I + np.pi/2

Vc = I * Xc
theta_Vc = theta_I - np.pi/2

Et = np.sqrt(2) * E * np.sin(w*t)
VRt = np.sqrt(2) * VR * np.sin(w*t + theta_VR)
VLt = np.sqrt(2) * VL * np.sin(w*t + theta_VL)
Vct = np.sqrt(2) * Vc * np.sin(w*t + theta_Vc)
fig,ax=plt.subplots()
ax.plot(t, Et, t,VRt, t,VLt, t,Vct)
```
