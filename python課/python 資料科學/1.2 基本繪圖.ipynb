## 7. 範例 -- 靜止物體做加速運動的速度與位置

v=at

$s=0.5at^2$

```python
import matplotlib.pyplot as plt
a = 1
T=[]
v=[]
s=[]
N = 50 #畫圖點數
Te = 5 #總運動時間
for i in range(N+1):
    t = i * Te/N #第 i 個點的時間
    T.append(t) 
    v.append(a*t)
    s.append(0.5*a*t*t)
fig, ax = plt.subplots()
ax.set(xlabel='t (sec)',
       ylabel='speed',
       title='Motion speed')
ax.plot(T, v,marker='*',label='Speed')

fig2, ax2 = plt.subplots()
ax2.set(xlabel='t (sec)',
       ylabel='distance',
       title='Motion distance')
ax2.plot(T, s,marker='o',label='Distance')



```

## 8. 練習 -- 從地面斜上拋射物體的速度與位置

<img align="center" style="padding-right:10px;" src="figures\projection.jpg">

v: 拋射初速

$\theta: 拋射角度(與地面夾角)$

$v_x (0) = v \cos(\theta)$

$v_y (0) = v \sin(\theta)$

$v_x (t) = v_x (0) $

$v_y (t) = v_y (0) - gt$

$x (t) = v_x (0) t$

$y (t) = v_y (0) t - 0.5gt^2$

$飛行時間 T_e = \frac{2 v_y (0)}{g} $

$角度 與  弳度 轉換$

$ 180^o =  \pi $

$弳度 = \frac{\pi}{180} 角度 $

```python
import matplotlib.pyplot as plt
import math #載入數學模組
g = 9.8

V0 = 100
angle = 30
Vx0 = V0*math.cos(math.pi/180*angle)
Vy0 = V0*math.sin(math.pi/180*angle)
T=[]
Vx = []
Vy = []
x = []
y = []
Te = 2*Vy0 / g #總飛行時間
N = 50 #畫圖點數
for i in range(N+1): 
    t = i * Te/N  #第 i 個點的時間
    T.append(t)
    Vx.append(Vx0)
    Vy.append(Vy0-g*t)
    x.append(Vx[i]*t)
    y.append(Vy0*t - 0.5*g*t*t)
    

fig, ax = plt.subplots()
ax.set(xlabel='t (sec)',
       ylabel='speed (m)',
       title='Motion speed')
ax.plot(T, Vx,marker='*',label='Speed x')
ax.plot(T, Vy,marker='*',label='Speed y')

ax.legend()

fig2, ax2 = plt.subplots()
ax2.set(xlabel='t (sec)',
       ylabel='Displacement (m)',
       title='Displacement')
ax2.plot(T, x,marker='*',label='Displacement x')
ax2.plot(T, y,marker='*',label='Displacement y')

ax2.legend()

fig3, ax3 = plt.subplots()
ax3.set(xlabel='Displacement (m)',
       ylabel='height (m)',
       title='Motion trajectory')
ax3.plot(x, y,marker='*',label='Speed')

```
