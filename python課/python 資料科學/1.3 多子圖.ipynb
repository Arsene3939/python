# 建立多子圖

## 1.使用 plt.subplots()

### 指令格式:

    plt.subplots( nrows=1, ncols=1, **fig_kw)
    
### 功能:

    建立1個圖形(figure) 與 1組子圖(axes,subplots).
    
### 範例

    plt.subplots( nrows=2, ncols=3)

<img align="center" style="padding-right:10px;" src="figures\plot_subplots.png">

### 1.1 多行多列

```python
import matplotlib.pyplot as plt
import math #載入數學模組
g = 9.8

V0 = 100
angle = 30
Vx0 = V0*math.cos(math.pi/180*angle)
Vy0 = V0*math.sin(math.pi/180*angle)
T=[]
Vx = []
Vy = []
x = []
y = []
Te = 2*Vy0 / g #總飛行時間
N = 50 #畫圖點數
for i in range(N+1): 
    t = i * Te/N  #第 i 個點的時間
    T.append(t)
    Vx.append(Vx0)
    Vy.append(Vy0-g*t)
    x.append(Vx[i]*t)
    y.append(Vy0*t - 0.5*g*t*t)
    

fig, ax = plt.subplots(nrows=2, ncols=2,figsize=(10,10))
ax[0][0].set(xlabel='t (sec)',
       ylabel='speed (m)',
       title='Motion speed')
ax[0][0].plot(T, Vx,marker='*',label='Speed x')
ax[0][0].plot(T, Vy,marker='*',label='Speed y')

ax[0][0].legend()


ax[0][1].set(xlabel='t (sec)',
       ylabel='Displacement (m)',
       title='X - Displacement')
ax[0][1].plot(T, x,marker='*',label='Displacement x')

ax[1][0].set(xlabel='t (sec)',
       ylabel='Displacement (m)',
       title='Y - Displacement')
ax[1][0].plot(T, y,marker='*',label='Displacement y')


ax[1][1].set(xlabel='Displacement (m)',
       ylabel='height (m)',
       title='Motion trajectory')
ax[1][1].plot(x, y,marker='*',label='Speed')
```

## 2. 使用 fig.add_subplot()

### 語法: 

    fig.add_subplot(nrows, ncols, index, **kwargs)
    
### 功能:

    建立1個圖形(figure) 與 1組子圖(axes,subplots).
    
### 範例

    ax1 = fig.add_subplot(231) # row-col-num
    ax2 = fig.add_subplot(232)
    ax3 = fig.add_subplot(233)
    ax4 = fig.add_subplot(234) 
    ax5 = fig.add_subplot(235)
    ax6 = fig.add_subplot(236)

<img align="center" style="padding-right:10px;" src="figures\plot_add_subplot.png">
   

```python
import matplotlib.pyplot as plt
import math #載入數學模組
g = 9.8

V0 = 100
angle = 30
Vx0 = V0*math.cos(math.pi/180*angle)
Vy0 = V0*math.sin(math.pi/180*angle)
T=[]
Vx = []
Vy = []
x = []
y = []
Te = 2*Vy0 / g #總飛行時間
N = 50 #畫圖點數
for i in range(N+1): 
    t = i * Te/N  #第 i 個點的時間
    T.append(t)
    Vx.append(Vx0)
    Vy.append(Vy0-g*t)
    x.append(Vx[i]*t)
    y.append(Vy0*t - 0.5*g*t*t)
    

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(221) 
ax2 = fig.add_subplot(222)
########################## 注意 ##########################
ax3 = fig.add_subplot(2,1,2) #此圖使用 2列1行，圖形寬度會變大

ax1.set(xlabel='t (sec)',
       ylabel='speed (m)',
       title='Motion speed')
ax1.plot(T, Vx,marker='*',label='Speed x')
ax1.plot(T, Vy,marker='*',label='Speed y')

ax1.legend()


ax2.set(xlabel='t (sec)',
       ylabel='Displacement (m)',
       title='X - Displacement')
ax2.plot(T, x,marker='*',label='Displacement x')

ax2.plot(T, y,marker='*',label='Displacement y')
ax2.legend()

ax3.set(xlabel='Displacement (m)',
       ylabel='height (m)',
       title='Motion trajectory')
ax3.plot(x, y,marker='*',label='Speed')
```

## 3. 使用 fig.add_axes( )


### 指令格式:
    
    fig.add_axes([left, bottom, width, height])

    所有參數皆為小於1的正小數，座標位置(0,0)位於圖形的左下角。寬度與高度的最大值為 1。


### 使用範例

    fig.add_axes([0.65, 0.65, 0.2, 0.2])

<img align="center" style="padding-right:10px;" src="figures\plot_add_axes.png">




```python
import matplotlib.pyplot as plt
import math #載入數學模組
g = 9.8

V0 = 100
angle = 30
Vx0 = V0*math.cos(math.pi/180*angle)
Vy0 = V0*math.sin(math.pi/180*angle)
T=[]
Vx = []
Vy = []
x = []
y = []
Te = 2*Vy0 / g #總飛行時間
N = 50 #畫圖點數
for i in range(N+1): 
    t = i * Te/N  #第 i 個點的時間
    T.append(t)
    Vx.append(Vx0)
    Vy.append(Vy0-g*t)
    x.append(Vx[i]*t)
    y.append(Vy0*t - 0.5*g*t*t)
    

fig = plt.figure(figsize=(10,10))

ax1 = fig.add_axes([0.05, 0.55, 0.25, 0.45], frameon=True)
ax2 = fig.add_axes([0.4, 0.55, 0.55, 0.45], frameon=True)
ax3 = fig.add_axes([0.05, 0.05, 0.9, 0.45], frameon=True)

ax1.set(xlabel='t (sec)',
       ylabel='speed (m)',
       title='Motion speed')
ax1.plot(T, Vx,marker='*',label='Speed x')
ax1.plot(T, Vy,marker='*',label='Speed y')

ax1.legend()


ax2.set(xlabel='t (sec)',
       ylabel='Displacement (m)',
       title='X - Displacement')
ax2.plot(T, x,marker='*',label='Displacement x')

ax2.plot(T, y,marker='*',label='Displacement y')
ax2.legend()

ax3.set(xlabel='Displacement (m)',
       ylabel='height (m)',
       title='Motion trajectory')
ax3.plot(x, y,marker='*',label='Speed')
```
