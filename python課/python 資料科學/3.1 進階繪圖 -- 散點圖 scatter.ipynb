# 3. 進階繪圖

## 3.1 散點圖 scatter

```python
import matplotlib.pyplot as plt
import numpy as np

x = [10, 20, 30, 40 ,50]
y = [30, 20, 40, 60, 70]

fig, ax = plt.subplots()
ax.scatter(x,y, s=1500, marker='*', c='red', alpha = 0.15)
```

### 函數原型
<pre>
ax.scatter(
    ['x', 'y', 's=None', 'c=None', 'marker=None', 'cmap=None', 'norm=None', 
     'vmin=None', 'vmax=None', 'alpha=None', 'linewidths=None', 'verts=None', 
     'edgecolors=None', '*', 'data=None', '**kwargs'],
)

x, y : 資料位置，x,y的長度需相同 

s : 散點尺寸，可為純量或與資料等長的陣列

c : 顏色，可為純量或與資料等長的陣列    

marker : 標詩外觀
</pre>

###  標誌代碼

```python
import matplotlib.pyplot as plt
import numpy as np

x = [10, 20, 30, 40 ,50]
y = [30, 20, 40, 60, 70]

size=[350,480,130,180,250]
color=['red','cyan','m','k','g']


fig, ax = plt.subplots()
ax.scatter(x,y, s=size, marker='*', c=color, alpha = 0.8)
```
