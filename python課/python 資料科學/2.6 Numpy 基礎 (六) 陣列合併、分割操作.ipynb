## 聲音操作--基本音符製作

產生低音的 Do、Re、...、Si等七個音的資料，取樣頻率11025赫茲，長度1秒，將其拼接成一筆，試著播放聲音。
<img align="center" src="figures\musicalFrequency.png">

```python
import numpy as np
import matplotlib.pyplot as plt
from Audio import Audio #載入聲音播放程式

samplingRate = 11025  #取樣頻率
M = 1  #聲音秒數
totalPoints = int(samplingRate * M) #總點數
t = np.linspace(0,M,totalPoints) #將 M 秒均分

Do = 30000*np.sin(2*np.pi*262*t)
Re = 30000*np.sin(2*np.pi*294*t)
Mi = 30000*np.sin(2*np.pi*330*t)

data = np.hstack((Do,Re,Mi))
print(Do.shape)
print(data.shape)
plt.plot(data[:100]) #取前1000筆資料
plt.show()
Audio(data,rate=samplingRate)
```

## 歌曲製作 
<img align="center" src="figures\beeSong.jpg">

```python
import numpy as np
from Audio import Audio #載入聲音播放程式

samplingRate = 11025  #取樣頻率
M = 2  #聲音秒數
totalPoints = int(samplingRate * M) #總點數
t = np.linspace(0,M,totalPoints) #將 M 秒均分

Do = 10000*np.sin(2*np.pi*262*t)
Re = 10000*np.sin(2*np.pi*294*t)
Mi = 10000*np.sin(2*np.pi*330*t)

L4 = int(0.5 * samplingRate)  #資料點數，設拍長為0.5秒
L2 = L4 * 2
L8 = int(L4/2)

data = np.hstack((Do[:L8],Do[:L8]*0,Re[:L8],Re[:L8]*0,Mi[:L4]))

Audio(data,rate=samplingRate)
```
