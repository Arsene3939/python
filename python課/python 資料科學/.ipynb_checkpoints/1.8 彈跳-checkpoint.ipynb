# 1.8 彈跳

## 上拋彈跳(無損彈跳)

```python
import math
import matplotlib.pyplot as plt
vo = 100  #物體初速，向上為正，向下為負
to = 0    #參考時間點
so = 0    #初始位置，在地面時其值為0，正值表示在空中。
g = 9.8   #重力加速度
T=[]      #時間資料串列
V=[]      #速度資料串列
S =[]     #位置資料串列
N = 5000  #總時間取樣點數
dt = 100/N #單位時間
for i in range(N):
    t = dt*i  #第 i 個時刻
    v=vo - g*(t-to)
    s=so + vo*(t-to)- 0.5*g*(t-to)**2
    if s<0 : #物體觸地時
        vo = -v  #運動方向轉向
        to=t     #重設參考時間
        so=0     #重設參考位置
    T.append(t)
    V.append(v)
    S.append(s)
fig,ax=plt.subplots(figsize=(15,6))
ax.plot(T,V, T,S)
```
