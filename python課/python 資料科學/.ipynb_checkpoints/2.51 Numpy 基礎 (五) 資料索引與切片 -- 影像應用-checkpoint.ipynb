## 影像切割

```python
import matplotlib.pyplot as plt
import numpy as np


img=plt.imread('figures\insect1.png')
print(img.shape)
plt.gray()
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(16,16))
img1=img.copy()
img1[:,0:180,:]=img[:,100:280,:]
img1[:,180:360,:]=img[:,280:100:-1,:]

ax[0,0].imshow(img)
ax[0,1].imshow(img1)
ax[1,0].imshow(img1[::-1,:,:])
ax[1,1].imshow(img1[:,:200,:])
```

### 分層控級

```python
import matplotlib.pyplot as plt
import numpy as np


img=plt.imread('figures\insect1.png')
print(img.shape)
plt.gray()

#漸層 -- 調透明度--第3層
img1 = img.copy()
for i in range(395):
    img1[:,i,3] = i/395
#img1[:,0:200,3] = 0.25

#減少緣色 -- 第1層
img2 = img.copy()
beta = 0.2
img2[:,:,1] = img2[:,:,1] * beta + np.zeros((297,395))*(1-beta)

#增強紅色 -- 第0層
img3 = img.copy()
img3[:,:,0] = img3[:,:,0] * beta + np.ones((297,395))*(1-beta)

fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(16,16))
ax[0][0].imshow(img)
ax[0][1].imshow(img1)
ax[1][0].imshow(img2)
ax[1][1].imshow(img3)
```

### 合層處理

```python
import matplotlib.pyplot as plt
import numpy as np


img=plt.imread('figures\insect1.png')
print(img.shape)
plt.gray()

#漸層 -- 調透明度--第3層
img1 = img.copy()*2
img1 = np.clip(img1,0,1)


#減少 x * beta + 0 * (1-beta)
img2 = img.copy()
beta = 0.8
img2 = img2 * beta + np.zeros(img2.shape)*(1-beta)

#增強 x * beta + 1 * (1-beta)
img3 = img.copy()
img3 = img3 * beta + np.ones(img3.shape)*(1-beta)
img3 = np.clip(img3,0,1)


fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(16,16))
ax[0][0].imshow(img)
ax[0][1].imshow(img1)
ax[1][0].imshow(img2)
ax[1][1].imshow(img3)
```
