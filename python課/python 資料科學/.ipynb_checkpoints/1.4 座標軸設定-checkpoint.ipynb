### 設定座標刻度、符號與極限

```python
import matplotlib.pyplot as plt
import math

fig,ax = plt.subplots()
T = []
Y1 = []
Y2 = []
for i in range(-100,101):
    t = i /50 * math.pi 
    T.append(t)
    Y1.append(math.sin(2*t))
    Y2.append(1.5*math.cos(t))
    

ax.plot(T, Y1, label="$sin(2t)$")
ax.plot(T, Y2, label="$1.5 cos(0.8t)$")
ax.legend(loc='best')

ax.set_xticks( [-6.28, -3.14, 0, 3.14, 6.28])
ax.set_xticklabels([r'$-2\pi$', r'$-\pi$', 0, r'$+\pi$', r'$+2\pi$'])
ax.set_yticks([-2, -1, 0, +1, 2])
ax.set_xlim([-6.3,6.3])
```

## 設定座軸位置

```python
import matplotlib.pyplot as plt
import math

fig,ax = plt.subplots()
T = []
Y1 = []
Y2 = []
for i in range(-100,101):
    t = i /50 * math.pi 
    T.append(t)
    Y1.append(math.sin(2*t))
    Y2.append(1.5*math.cos(t))
    

ax.plot(T, Y1, label="$sin(2t)$")
ax.plot(T, Y2, label="$1.5 cos(0.8t)$")
ax.legend(loc='upper left')

ax.set_xticks( [-6.28, -3.14, 0, 3.14, 6.28])
ax.set_xticklabels([r'$-2\pi$', r'$-\pi$', 0, r'$+\pi$', r'$+2\pi$'])
ax.set_yticks([-2, -1, 0, +1, 2])
ax.set_xlim([-6.3,6.3])

ax.spines['left'].set_position('zero')   # makee x and y axes go through
ax.spines['bottom'].set_position('zero') # the origin
ax.spines['right'].set_color('none')     # hide the unnecessary
ax.spines['top'].set_color('none')       # spines ("the box" around the plot)
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
X = []
Y1 = []
Y2 = []
for i in range(-200,201):
    x = i /50  
    X.append(x)
    y1 = x**2 - 4
    y2 = x**2-x-6
    Y1.append(y1)
    Y2.append(y2)
    

ax.plot(X, Y1, label="$x^2 - 4$")
ax.plot(X, Y2, label="$x^2 - x - 6$")
ax.legend(loc='upper left')

ax.spines['left'].set_position('zero')   # makee x and y axes go through
ax.spines['bottom'].set_position('zero') # the origin
ax.spines['right'].set_color('none')     # hide the unnecessary
ax.spines['top'].set_color('none')       # spines ("the box" around the plot)
```
