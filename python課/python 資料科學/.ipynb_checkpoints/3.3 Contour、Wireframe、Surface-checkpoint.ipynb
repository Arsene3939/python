## 範例 
$z=\sqrt{10x^2 + y^2 }$

## Contour

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

fig,ax = plt.subplots(figsize=(10,8))
start, stop, n_values = -8, 8, 800

x = np.linspace(start, stop, n_values)
y = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x, y)


Z = np.sqrt(10*X**2 + Y**2)
ax.contour(X, Y, Z)
ax.set_title('Contour Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
```

## ContourFill

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

fig,ax = plt.subplots(figsize=(10,8))
start, stop, n_values = -8, 8, 800

x = np.linspace(start, stop, n_values)
y = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x, y)


Z = np.sqrt(10*X**2 + Y**2)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp, ax=ax, orientation='horizontal')
ax.set_title('Contour Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')

```

## Wireframe

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d")

start, stop, n_values = -8, 8, 800

x = np.linspace(start, stop, n_values)
y = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x, y)


Z = np.sqrt(10*X**2 + Y**2)
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title('wireframe');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

## Surface

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d")

start, stop, n_values = -8, 8, 800

x = np.linspace(start, stop, n_values)
y = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x, y)


Z = np.sqrt(10*X**2 + Y**2)
ax.set_title('surface');
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X, Y, Z)#
#, rstride=1, cstride=1, cmap='winter', edgecolor='none')
```

## 範例: 
### $ 平面上 (x_0, y_0 )位置有一電荷Q，畫出此電荷周圍的電位圖$
電位公式
$V(x,y)=k\cfrac{Q}{r} = k\cfrac{Q}{\sqrt{(x-x_0 )^2 + (y- y_0)^2}}$

### Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

r = np.arange(0.2,10,0.1)

Q = 100E-9
k = 9E9
V = k*Q/r

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(r,V)
#ax.set_title('V')
ax.set(xlabel='r',ylabel='V',title='Voltage potential')
```

### Contour

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.arange(-10.5,10,0.2)
y = np.arange(-10.5,10,0.2)
X,Y = np.meshgrid(x,y)

Q = 100E-9
x0 = 5
y0 = 0
r = ((X-x0)**2 + (Y-y0)**2)**0.5
k = 9E9
V = k*Q/r

fig, ax = plt.subplots(figsize=(8,6))
cp = ax.contourf(X,Y,V)
fig.colorbar(cp, ax=ax, orientation='horizontal')
ax.set_title('V')
```

### Wireframe

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.arange(-10.5,10,0.2)
y = np.arange(-10.5,10,0.2)
X,Y = np.meshgrid(x,y)

Q = 100E-9
x0 = 5
y0 = 0
r = ((X-x0)**2 + (Y-y0)**2)**0.5
k = 9E9
V = k*Q/r
V = np.clip(V,-2000,2000)

from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, V, color='green')
ax.set_title('V')
```

### Surface

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.arange(-10.5,10,0.2)
y = np.arange(-10.5,10,0.2)
X,Y = np.meshgrid(x,y)

Q = 100E-9
x0 = 5
y0 = 0
r = ((X-x0)**2 + (Y-y0)**2)**0.5
k = 9E9
V = k*Q/r

V = np.clip(V,-2000,2000)
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, V)
ax.set_title('V')
```

## 練習: 
### $ 平面上 (x_1, y_1 )與(x_2, y_2 )位置各有一電荷Q_1 與 Q_2，畫出電荷周圍的電位圖$
電位公式
$V(x,y)=k\cfrac{Q_1 }{r_1}+ k\cfrac{Q_2 }{r_2} = k\cfrac{Q_1 }{\sqrt{(x-x_1 )^2 + (y- y_1)^2}}+ k\cfrac{Q_2 }{\sqrt{(x-x_2 )^2 + (y- y_2)^2}}$

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.arange(-10.5,10,0.2)
y = np.arange(-10.5,10,0.2)
X,Y = np.meshgrid(x,y)

Q1 = 100E-6
Q2 = -100E-6
x1 = 5; y1=0;
x2 = -5; y2=0;

r1 = ((X-x1)**2 + (Y-y1)**2)**0.5
r2 = ((X-x2)**2 + (Y-y2)**2)**0.5
k = 9E9
V = k*Q1/r1  + k*Q2 /r2

fig, ax = plt.subplots(figsize=(8,6))
cp = ax.contourf(X,Y,V)
fig.colorbar(cp, ax=ax, orientation='horizontal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Contour of V');
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.arange(-10.5,10,0.2)
y = np.arange(-10.5,10,0.2)
X,Y = np.meshgrid(x,y)

Q1 = 100E-9
Q2 = -100E-9
x1 = 5; y1=0;
x2 = -5; y2=0;

r1 = ((X-x1)**2 + (Y-y1)**2)**0.5
r2 = ((X-x2)**2 + (Y-y2)**2)**0.5

k = 9E9
V = k*Q1/r1  + k*Q2 /r2


#V = np.clip(V,-2000,2000)


fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d")

ax.plot_wireframe(X, Y, V, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Wireframe of V');

```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, V, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
ax.set_title('Surface plot of V');
```
