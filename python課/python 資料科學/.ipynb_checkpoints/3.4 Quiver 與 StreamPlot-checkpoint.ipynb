# 3.4 Quiver 與 StreamPlot

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

x_pos = np.array([0, 2, 4, 6])
y_pos = np.array([0, 0, 0, 0])
x_direct = np.array([1, 1, 1, 1])
y_direct = np.array([1, 1, 1, 1])

ax.quiver(x_pos,y_pos,x_direct,y_direct)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(ncols=2,figsize=(10,6))

x = np.array([0, 1])
y = np.array([0, 1])
X,Y = np.meshgrid(x,y)
x_direct = np.array([[1,1],
                     [1,1]])
y_direct = np.array([[1,1],
                     [1,1]])

ax[0].quiver(X,Y,x_direct,y_direct)
ax[1].streamplot(X,Y,x_direct,y_direct,density=0.5)

ax[0].axis([-0.1,1.1,-0.1,1.1])
ax[1].axis([-0.1,1.1,-0.1,1.1])
```

## 補充: 平面上的方向向量與其單位向量

$平面上由(x_1, y_1 )到(x_2, y_2 )的方向向量可表示為$
$$\vec{r}=(x_2 - x_1, y_2 - y_1 )=(r_x , r_y)$$
$向量 \vec{r}的長度可表示為$
$$r=\sqrt{(x_2 -x_1 )^2 + (y_2- y_1)^2}=\sqrt{{r_x}^2 + {r_y}^2}$$
$向量 \vec{r}的單位向量\hat{r}，其長度為1，可表示為$
$$ \hat{r} = \cfrac{\vec{r}}{r} = \cfrac{(r_x , r_y)}{r}=(\cfrac{r_x}{r},\cfrac{r_y}{r})$$

## 範例: 
$平面上由(1, 1 )到(4, 5 )的方向向量可表示為$
$$\vec{r}=(4 - 1, 5 - 1 )=(3,4)$$
$向量 \vec{r}的長度可表示為$
$$r=\sqrt{(4 -1 )^2 + (5- 1)^2} = 5$$
$向量 \vec{r}的單位向量\hat{r}可表示為$
$$ \hat{r} = \cfrac{\vec{r}}{r} = \cfrac{(3, 4)}{5}=(0.6, 0.8)$$

## 電場的向量表示式
$ 平面上 (x_0, y_0 )位置有一電荷Q，畫出此電荷周圍的電場表示式為$

$$E(x,y)=k\cfrac{Q}{r^2}\hat{r}=k\cfrac{Q}{r^3}\vec{r}  =(k\cfrac{Qr_x}{r^3},k\cfrac{Qr_y}{r^3})$$

## 直線圖

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

r = np.arange(0.2,10,0.1)

Q = 100E-9
k = 9E9
E = k*Q/r**2

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(r,E)
ax.set(xlabel='r',ylabel='E',title='E field intensity plot')
```

## Quiver

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3,3,0.2)
y = np.arange(-3,3,0.2)
X,Y = np.meshgrid(x,y)

Q1 = 1E-9
x1 = 0; y1 = 0
r1 = ((X-x1)**2 + (Y-y1)**2)**0.5
k = 9E9
Ex = k*Q1*(X-x1)/r1**3 
Ey = k*Q1*(Y-y1)/r1**3 

M = 10
idx=Ex>M; Ex[idx]=M; idx=Ex<-M; Ex[idx]=-M
idx=Ey>M; Ey[idx]=M; idx=Ey<-M; Ey[idx]=-M

fig, ax = plt.subplots(figsize=(6,6))
ax.quiver(X,Y,Ex,Ey)
ax.set_aspect('equal')
ax.scatter(x1,y1, s=100, c='r')

ax.set_title('Stream Plot of a Point Charges')
```

## StreamPlot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)


Q1 = 1E-6
x1 = 0; y1 = 0
r1 = ((X-x1)**2 + (Y-y1)**2)**0.5
k = 9E9
Ex = Q1*(X-x1)/r1**3 
Ey = Q1*(Y-y1)/r1**3 

fig, ax = plt.subplots(figsize=(6,6))
ax.streamplot(X,Y,Ex,Ey, density = 2)
ax.set_aspect('equal')
ax.scatter(x1,y1, s=100, c='r')

ax.set_title('Stream Plot of a Point Charges')
a=0
```

## 2個電荷的電場圖

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)

Q1 = 10E-6
Q2 = -10E-6
x1 = 2; y1 = 0
x2 =-2; y2 = 0

r1 = ((X-x1)**2 + (Y-y1)**2)**0.5
r2 = ((X-x2)**2 + (Y-y2)**2)**0.5

k = 9E9
Ex = k*Q1*(X-x1)/r1**3 + k*Q2*(X-x2)/r2**3
Ey = k*Q1*(Y-y1)/r1**3 + k*Q2*(Y-y2)/r2**3

fig, ax = plt.subplots(figsize=(6,6))
ax.streamplot(X,Y,Ex,Ey, density = 2)
ax.set_aspect('equal')
ax.scatter(x1,y1, s=100, c='r')
ax.scatter(x2,y2, s=100, c ='b')
ax.set_title('Stream Plot of Two Point Charges')
```
