# RL 電路圖與公式

<img align="center" style="padding-right:10px;" src="figures\RL_Circuit.gif">

$\omega = 2 \pi f$   
$e(t)=\sqrt{2}E\sin(\omega t + \theta_E)$  
$\vec{E} = E \angle \theta_E $  

$\vec{X_L}=X_L\angle \theta_L, X_L = \omega L, \theta_L = 90°$  
$\vec{Z} = Z \angle \theta_Z,  Z=\sqrt{R^2 + X_L^2}, \theta_L = \arctan \cfrac{X_L}{R}$  
$\vec{I} = \cfrac{\vec{E}}{\vec{Z}} = I \angle \theta_I, I = \cfrac{E}{Z}, \theta_I = \theta_E - \theta_Z$  
$\vec{V_R} =\vec{I} R =  V_R \angle \theta_{V_R}, V_R = IR, \theta_{V_R} = \theta_I$  
$\vec{V_L} =\vec{I} \vec{X_L} = V_L \angle \theta_{V_L}, V_L = IX_L , \theta_{V_L} = \theta_I + 90°$


$e(t)=\sqrt{2}E\sin(\omega t + \theta_E)$  
$i(t)=\sqrt{2}I\sin(\omega t + \theta_I)$  
$v_R (t)=\sqrt{2}V_R \sin(\omega t + \theta_{V_R})$  
$v_L (t)=\sqrt{2}V_L \sin(\omega t + \theta_{V_L})$  

## 單一頻率 阻抗、電壓、電流計算

```python
import math
import matplotlib.pyplot as plt
R = 4
L = 3E-3
E = 10
theta_E = 0
#f = 1000
#w = 2*math.pi*f
w = 1000
f = w /(2*math.pi)

#信號週期
Ts = 1 / f
print('頻率:%f 週期:%f' % (f,Ts))

XL = w * L
Z = (R**2 + XL**2)**0.5
theta_Z = math.atan(XL/R)

I = E / Z
theta_I = theta_E - theta_Z

VR = I * R
theta_VR = theta_I

VL = I * XL
theta_VL = theta_I + math.pi/2

print('R=%5.1f,  XL=%5.1f' % (R,XL))
print('Z=%5.1f,  theta_Z=%5.1f' % (Z, theta_Z*180/math.pi))
print('I=%5.1f,  theta_I=%5.1f' % (I,theta_I*180/math.pi))
print('VR=%5.1f, theta_VR=%5.1f' % (VR,theta_VR*180/math.pi))
print('VL=%5.1f, theta_VL=%5.1f' % (VL,theta_VL*180/math.pi))
```

## 單一頻率 電壓、電流波形圖

```python

#畫圖點數
N = 200
dt = 2 * Ts/N  #畫2個週期
T = [dt*i for i in range(N)]
Et = [1.414*E*math.sin(w*t) for t in T]
VRt = [1.414*VR*math.sin(w*t + theta_VR) for t in T]
VLt = [1.414*VL*math.sin(w*t + theta_VL) for t in T]

fig,ax = plt.subplots()
ax.plot(T,Et,label='Et')
ax.plot(T,VRt,label='VRt')
ax.plot(T,VLt,label='VLt')
ax.legend()
```

## 單一頻率 電壓相量圖


```python
fig = plt.figure()
ax = plt.subplot(111, polar=True)
ax.set_rmax(20)
#line, = ax.plot(theta_VR, VR,theta_VL, VL,theta_VC, VC )
line1, = ax.plot([theta_E,0],[E,0],label='E')
line2, = ax.plot([theta_VR,0],[VR,0],label='VR')
line3, = ax.plot([theta_VL,0],[VL,0],label='VL')


ax.legend(loc='center left')
```

## RLC 電路圖與公式

$\omega = 2 \pi f$   
$e(t)=\sqrt{2}E\sin(\omega t + \theta_E)$  
$\vec{E} = E \angle \theta_E $  

$\vec{X_L}=X_L\angle \theta_L, X_L = \omega L, \theta_L = 90°$  
$\vec{X_C}=X_C\angle \theta_C, X_C = \cfrac{1}{\omega C}, \theta_C = -90°$  
$\vec{X} = \vec{X_L} - \vec{X_C} =X\angle \pm 90°$  
$\vec{Z} = Z \angle \theta_Z,  Z=\sqrt{R^2 + X^2}, \theta_L = \arctan \cfrac{X}{R}$  
$\vec{I} = \cfrac{\vec{E}}{\vec{Z}} = I \angle \theta_I, I = \cfrac{E}{Z}, \theta_I = \theta_E - \theta_Z$  
$\vec{V_R} =\vec{I} R =  V_R \angle \theta_{V_R}, V_R = IR, \theta_{V_R} = \theta_I$  
$\vec{V_L} =\vec{I} \vec{X_L} = V_L \angle \theta_{V_L}, V_L = IX_L , \theta_{V_L} = \theta_I + 90°$  
$\vec{V_C} =\vec{I} \vec{X_C} = V_C \angle \theta_{V_C}, V_C = IX_C , \theta_{V_C} = \theta_I - 90°$


$e(t)=\sqrt{2}E\sin(\omega t + \theta_E)$  
$i(t)=\sqrt{2}I\sin(\omega t + \theta_I)$  
$v_R (t)=\sqrt{2}V_R \sin(\omega t + \theta_{V_R})$  
$v_L (t)=\sqrt{2}V_L \sin(\omega t + \theta_{V_L})$  
$v_C (t)=\sqrt{2}V_C \sin(\omega t + \theta_{V_C})$
